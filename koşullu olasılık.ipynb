```python
# Probabilities
Probability_H = 0.20  # Probability that a borrower is from the high-risk group
Probability_L = 0.80  # Probability that a borrower is from the low-risk group

P_D_given_H = 0.10  # Probability of default given high-risk group
P_D_given_L = 0.02  # Probability of default given low-risk group

# Total probability of defaulting (P(D))
P_D = (P_D_given_H * Probability_H) + (P_D_given_L * Probability_L)

# Conditional probability P(H|D) using Bayes' Theorem
P_H_given_D = (P_D_given_H * Probability_H) / P_D

print(f"The probability that a borrower is from the high-risk group given that they have defaulted is {P_H_given_D:.4f}")
```

**Problem:**

A financial analyst is assessing the risk of default on loans given to two groups of borrowers: high-risk and low-risk. Historical data shows that 20% of the borrowers are in the high-risk group and 80% are in the low-risk group. The default rates are 10% for the high-risk group and 2% for the low-risk group. If a randomly selected borrower has defaulted on their loan, what is the probability that this borrower is from the high-risk group?

```python
# Probabilities
P_H = 0.20  # Probability that a borrower is from the high-risk group
P_L = 0.80  # Probability that a borrower is from the low-risk group

P_D_given_H = 0.10  # Probability of default given high-risk group
P_D_given_L = 0.02  # Probability of default given low-risk group

# Total probability of defaulting (P(D))
P_D = (P_D_given_H * P_H) + (P_D_given_L * P_L)

# Conditional probability P(H|D) using Bayes' Theorem
P_H_given_D = (P_D_given_H * P_H) / P_D

print(f"The probability that a borrower is from the high-risk group given that they have defaulted is {P_H_given_D:.4f}")
```
