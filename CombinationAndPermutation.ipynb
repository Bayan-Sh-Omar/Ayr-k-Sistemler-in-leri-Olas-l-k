*   HeadPhone

*   Printer

I can  buy just three items according to the budget.

**Solution**

Five different items  n=5
i can buy three items r=3

A) How many different choices can i make?

Combination
C(n,r)= n!/(n-r)!*r!
C(5,3)=5!/(5-3)!*3!=10 Tawfiq can make 10 different choices

B) If the order is important in choosingextenranl items for Laptop, how many different choices can be made?

Permutation

P(n,r)= n!/(n-r)!
P(5,3)= 5!/(5-3)! =(5 x 4 x 3 x 2 x 1)/(2 x 1) ==> 5 x 4 x 3=60


```python
from types import new_class
from math import factorial
from itertools import permutations
from itertools import combinations


NumberOfItems=5
r=3

def MyCombinations(n,r):
    comb = combinations(['HeadPhone', 'Mouse', 'Keyboard', 'Printer', 'Scanner'], r)
    print("\nThe count of Combinations are ",factorial(n)/(factorial(r)*factorial(n-r)))
    for i in list(comb):
      print ("Comb :",i)
MyCombinations(NumberOfItems,r)

def MyPermutations(n,r):
    perm = permutations(['HeadPhone', 'Mouse', 'Keyboard', 'Printer', 'Scanner'], r)
    print("\nThe count of  Permutations)=",(factorial(n)/(factorial(n-r))))
    for i in list(perm):
      print ("Perm :",i)

MyPermutations(NumberOfItems,r)


```

```python
def permutations_from_scratch(elements):
    def helper(current_permutation):
        # If the current permutation length is equal to the length of elements, append it to the results
        if len(current_permutation) == len(elements):
            results.append(current_permutation[:])
            return

        # Iterate through the elements
        for i in range(len(elements)):
            # If the element is already in the current permutation, skip it
            if elements[i] in current_permutation:
                continue

            # Add the current element to the permutation
            current_permutation.append(elements[i])
            # Recursively call helper for the next element
            helper(current_permutation)
            # Backtrack by removing the last element added
            current_permutation.pop()

    # Initialize results list
    results = []
    # Start the recursive helper function with an empty list
    helper([])
    return results

# Example :
elements = ['HeadPhone', 'Mouse', 'Keyboard', 'Printer', 'Scanner']

# Generate permutations from scratch
permutations_list = permutations_from_scratch(elements)

# Print the permutations
for perm in permutations_list:
    print(perm)
```

```python
def combinations_from_scratch(elements, k):
    def helper(start, current_combo):
        # If the current combination length is equal to k, append it to the results
        if len(current_combo) == k:
            results.append(current_combo[:])
            return

        # Iterate through the elements starting from the current start index
        for i in range(start, len(elements)):
            # Add the current element to the combination
            current_combo.append(elements[i])
            # Recursively call helper for the next element
            helper(i + 1, current_combo)
            # Backtrack to remove the last element
            current_combo.pop()

    # Initialize results list
    results = []
    # Start the recursive helper function
    helper(0, [])
    return results

# Example :
elements = ['HeadPhone', 'Mouse', 'Keyboard', 'Printer', 'Scanner']
k = 3

# Generate combinations from scratch
combinations_list = combinations_from_scratch(elements, k)

# Print the combinations
for combo in combinations_list:
    print(combo)
```
